# Top Countries Participation

```python
Top_Countries =olympics.value_counts(["Country_Code"]).reset_index(name= "Counts")[:10]
Top_Countries
```

```python
sns.set_style('darkgrid')
matplotlib.rcParams['figure.figsize'] = (12, 10)

plt.title("Overall Participation Country-Wise")
barplot = sns.barplot(data=Top_Countries, x="Country_Code", y="Counts", hue="Country_Code", palette="colorblind", dodge=False)
barplot.set_xticklabels(rotation=50, labels=Top_Countries.Country_Code)
plt.ylabel("Counts")
plt.show()
```

# Gender Participation

```python
Gender_Counts =olympics.value_counts(["Gender"]).reset_index(name= "Counts")[:10]

#To get the percentage
Gender_Counts['Percentage %'] = round(Gender_Counts.Counts / Gender_Counts.Counts.sum() * 100, 2)

Gender_Counts
```

```python
Gender_percentage = Gender_Counts["Percentage %"].to_list()
plt.title('Gender Distribution', fontsize=30, loc='Right')
plt.pie(Gender_percentage, colors=['#AEC6CF', '#FFB6C1'], labels= ["Men", "Women"], radius=1.2, autopct= "%0.1f%%", pctdistance = 0.77,
        textprops = {"size" : 17, "color" : "#000000"}, explode= [0.00, 0.00])

centre_circle = plt.Circle((0,0), 0.8, fc = "w")
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.show()
```

Male seems to dominating in terms of participation.

-> To filter gender that are female and year

```python
female_participants = olympics[(olympics["Gender"] == "Women")][["Gender", "Year"]]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
sns.lineplot(data=female_participants, x='Year', y='Gender', color= "blue")
plt.title("Female Participation over the Years ")

plt.show()
```

From the above chart we conclude that female participation has increased significantly over the year.

# Medals won by different genders

```python
Gender_by_Medals = pd.crosstab(olympics["Gender"],olympics["Medal"])
Gender_by_Medals
```

```python
sns.set_style('darkgrid')
matplotlib.rcParams['figure.figsize'] = (12, 10)
plt.title("Overall Participation CountryWise")

#To moved the Medal to become the inner row index
stacked = Gender_by_Medals.stack().reset_index().rename(columns={0:'value'})
sns.barplot(x=stacked.Medal, y=stacked.value, hue=stacked.Gender, palette="inferno")
plt.ylabel("Counts of Medal")
plt.show()
```

# Year In Which India Won Its First Gold Medal

```python
# Making a new dataset which contains only the sports in which atleast one gold medal is won.
gold_dataset= olympics[olympics["Medal"]=="Gold"]
gold_dataset.head()
```

```python
First_Medal = gold_data[gold_data['Medal'] == 'Gold']
Firstmedal_India = First_Medal[First_Medal['Country_Code'] == 'IND']

lstt = []
for i in Firstmedal_India['Year'].unique():
     lstt.append(i)
min(lstt)
print(f"INDIA WON FIRST GOLD MEDAL IN SUMMER OLYMPICS: {min(lstt)}")
```

```python
India_goldmedals=gold_dataset[gold_dataset["Country_Code"]=="IND"]
India_goldmedals=India_goldmedals.sort_values(by="Year", ascending=True)
print(f"INDIA WON FIRST GOLD MEDAL IN SUMMER OLYMPICS:\n")
India_goldmedals.head(1)
```

```python
Country_by_medal = olympics[(olympics["Country_Code"] == "IND")][["Sport", "Medal"]]
Country_by_medal = Country_by_medal.value_counts(["Sport"]).reset_index(name = "Counts_of_Medal")[:10]
print(f"TOTAL MEDALS WON BY INDIA IN SUMMER OLYMPICS:\n")
Country_by_medal
```

```python

sports = ["Hockey", "Shooting", "Boxing", "Tennis", "Weightlifting", "Wrestling"]
medals = [16, 2, 1, 1, 1, 1]

colors = ['#FF9999','#66B2FF','#99FF99','#FFCC99','#C2C2F0','#FFB6C1']
explode = [0.1 if x == max(medals) else 0 for x in medals]

plt.figure(figsize=(8, 8))
plt.pie(
    medals,
    labels=sports,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    explode=explode,
    textprops={'fontsize': 14}
)
plt.title("India's Total Medals by Sport (Summer Olympics)", fontsize=16)
plt.axis('equal')
plt.show()
```

# Sport Which Has Most Female Gold Medalist (Top 5)
